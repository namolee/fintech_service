```python
# 빈 셀을 위의 값으로 채우기
data_code['대분류'] = data_code['대분류'].ffill()
data_code['중분류'] = data_code['중분류'].ffill()
data_code['소분류'] = data_code['소분류'].ffill()
data_code['세분류'] = data_code['세분류'].ffill()
data_code['세세분류'] = data_code['세세분류'].ffill()

# 필요한 열만 선택하여 보기 좋게 정리
data_code_cleaned = data_code[['대분류', '대분류명', '중분류', '중분류명', '소분류', '소분류명', '세분류', '세분류명', '세세분류', '세세분류명']]
data_code_cleaned.head()  # 정리된 결과 일부 확인
```
