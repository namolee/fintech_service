```python
import pandas as pd

# Load the four CSV files
file_paths = [
    './review_data/bangbang_makguksu_reviews_google(1).csv',
    './review_data/bangbang_makguksu_reviews_kakao(1).csv',
    './review_data/뱅뱅막국수_reviews_최종.csv'
]

# Read and combine the files into a single DataFrame
dataframes = [pd.read_csv(file_path) for file_path in file_paths]
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a CSV file
combined_df.to_csv('뱅뱅막국수_combined_reviews.csv', index=False)
```
