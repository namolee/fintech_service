```python
import requests
import pandas as pd
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

# API URL과 API 키 설정
base_url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
api_key = "ON71dnhxzcL78L3dMT1PwVM3kG/znbDB/iAXy7Xn2kI+XxoqZ6XUC3ki358VW3+7taYR3EKKeyFxejkdtEhYRw=="

# 필요한 컬럼 목록
columns_needed = [
    "bizesId", "bizesNm", "brchNm", "indsLclsCd", "indsLclsNm",
    "indsMclsCd", "indsMclsNm", "indsSclsCd", "indsSclsNm",
    "ksicCd", "ksicNm", "adongCd", "adongNm", "ldongCd",
    "ldongNm", "rdnmCd", "rdnm", "lon", "lat", "ctprvnCd"
]

# 특정 지역 데이터만 가져오는 함수 정의
def fetch_page(page_no):
    params = {
        "ServiceKey": api_key,
        "pageNo": page_no,
        "numOfRows": 100,
        "divId": "indsLclsCd",
        "key": "I2",
        "type": "json"
    }
    while True:
        try:
            response = requests.get(base_url, params=params, timeout=10)
            data = response.json()
            
            # 데이터가 없으면 None 반환 (더 이상 페이지가 없을 경우)
            if not data['body']['items']:
                return None
            
            # ctprvnCd가 41인 데이터만 선택
            page_data = [
                {col: item.get(col) for col in columns_needed}
                for item in data['body']['items'] if item.get("ctprvnCd") == "41"
            ]
            print(f"{page_no} 페이지 수집 완료")
            return page_data

        except requests.exceptions.Timeout:
            print(f"페이지 {page_no} 요청 시간이 초과되었습니다. 재시도합니다.")
            time.sleep(1)  # 재시도 전 1초 대기

# 수집할 총 페이지 수 지정 (예: 100 페이지)
total_pages = 10000
all_data = []

# ThreadPoolExecutor를 사용하여 병렬로 페이지 요청
with ThreadPoolExecutor(max_workers=10) as executor:
    future_to_page = {executor.submit(fetch_page, page_no): page_no for page_no in range(1, total_pages + 1)}
    for future in as_completed(future_to_page):
        result = future.result()
        if result:
            all_data.extend(result)

# DataFrame으로 변환
df = pd.DataFrame(all_data)
print(df.head())
```
