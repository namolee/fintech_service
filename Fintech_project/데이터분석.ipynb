```python
from sqlalchemy import create_engine
import pandas as pd

# 데이터베이스 연결 정보
db = "mysql"
dbtype = "pymysql"
user = "root"
password = "1234"
host = "127.0.0.1"
port = 3306
database = "project"  # 사용할 데이터베이스 이름

# SQLAlchemy 엔진 생성
engine = create_engine(f"{db}+{dbtype}://{user}:{password}@{host}:{port}/{database}")
conn = engine.connect()

try:
    # SQL 쿼리 실행
    query = "SELECT * FROM stores;"  # 테이블명 변경 필요
    data = pd.read_sql(query, conn)  # Pandas DataFrame으로 데이터 읽기

    # 데이터 확인
    print(data.head())  # 데이터의 첫 5행 출력
finally:
    # 연결 종료
    conn.close()
```

```python
import matplotlib.pyplot as plt

# 'price_range' 값이 존재하고 'category'가 '한식'인 데이터 필터링
filtered_data = data[(data['price_range'].notnull()) & (data['category'] == '한식')]

# 그래프를 그리기 위해 가게 이름과 가격대만 가져오기
store_names = filtered_data['store_name']  # 가게 이름 열 (예시)
price_ranges = filtered_data['price_range']

# 막대 그래프 그리기
plt.figure(figsize=(12, 6))
plt.bar(store_names, price_ranges, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Price Range for Korean Cuisine Stores (한식)', fontsize=14)
plt.xlabel('Store Name', fontsize=12)
plt.ylabel('Price Range', fontsize=12)
plt.xticks(rotation=45, ha='right')  # 가게 이름 회전
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 'price_range' 값이 존재하고 'category'가 '한식'인 데이터 필터링
filtered_data = data[(data['price_range'].notnull()) & (data['category'] == '한식')]

# 네이버, 구글, 카카오 평점의 평균 계산
filtered_data['rating_avg'] = filtered_data[['naver_rating', 'google_rating', 'kakao_rating']].mean(axis=1)

# 데이터 추출
x_values = filtered_data['rating_avg']  # X축: 평점 평균
y_values = filtered_data['price_range']  # Y축: 가격대
store_names = filtered_data['store_name']  # 점 옆에 표시할 가게 이름

# 산점도 그리기
plt.figure(figsize=(12, 6))
plt.scatter(x_values, y_values, color='blue', alpha=0.7)

# 점 옆에 가게 이름 표시
for i, store_name in enumerate(store_names):
    plt.text(x_values.iloc[i], y_values.iloc[i], store_name, fontsize=9, ha='right')

# 그래프 설정
plt.title('한식 가게별 가격대와 평점 평균', fontsize=14)
plt.xlabel('Average Rating (평점 평균)', fontsize=12)
plt.ylabel('Price Range (가격대)', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from adjustText import adjust_text

# 'price_range' 값이 존재하고 'category'가 '한식'인 데이터 필터링
filtered_data = data[(data['price_range'].notnull()) & (data['category'] == '한식')]

# 네이버, 구글, 카카오 평점의 평균 계산
filtered_data['rating_avg'] = filtered_data[['naver_rating', 'google_rating', 'kakao_rating']].mean(axis=1)

# 데이터 추출
x_values = filtered_data['rating_avg']  # X축: 평점 평균
y_values = filtered_data['price_range']  # Y축: 가격대
store_names = filtered_data['store_name']  # 점 옆에 표시할 가게 이름

# 산점도 그리기
plt.figure(figsize=(12, 6))
plt.scatter(x_values, y_values, color='blue', alpha=0.7)

# 점 옆에 가게 이름 표시를 위한 텍스트 객체 리스트
texts = []
for i, store_name in enumerate(store_names):
    texts.append(plt.text(x_values.iloc[i], y_values.iloc[i], store_name, fontsize=9))

# 텍스트 자동 위치 조정
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='gray', alpha=0.7))

# 그래프 설정
plt.title('한식 가게별 가격대', fontsize=14)
plt.xlabel('Average Rating (평점 평균)', fontsize=12)
plt.ylabel('Price Range (가격대)', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
