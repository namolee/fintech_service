```python
from konlpy.tag import Okt

okt = Okt()

# 명사와 형용사만 추출하는 함수
def extract_keywords(text):
    return [word for word, pos in okt.pos(text) if pos in ['Noun', 'Adjective']]

# 키워드를 형태소 분석하여 명사와 형용사만 추출
taste_keywords_morphs = [extract_keywords(keyword) for keyword in taste_keywords_negative]
price_keywords_morphs = [extract_keywords(keyword) for keyword in price_keywords_negative]
menu_keywords_morphs = [extract_keywords(keyword) for keyword in menu_keywords_negative]

# 리뷰가 어떤 카테고리에 해당하는지 판별하는 함수
def categorize_review(review):
    review_morphs = extract_keywords(review)  # 리뷰에서 명사와 형용사만 추출
    
    # '맛' 관련 키워드 비교
    if any(any(morph in review_morphs for morph in keyword) for keyword in taste_keywords_morphs):
        return '맛'
    
    # '가격' 관련 키워드 비교
    elif any(any(morph in review_morphs for morph in keyword) for keyword in price_keywords_morphs):
        return '가격'
    
    # '메뉴' 관련 키워드 비교
    elif any(any(morph in review_morphs for morph in keyword) for keyword in menu_keywords_morphs):
        return '메뉴'
    
    else:
        return '기타'  # 해당하는 키워드가 없을 경우

# 'Category' 컬럼 생성하여 부정적 리뷰의 카테고리를 표시
negative_reviews_df.loc[:, 'Category'] = negative_reviews_df['Review'].apply(categorize_review)
```
