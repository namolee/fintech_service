```python
import pandas as pd
from sqlalchemy import create_engine

# CSV íŒŒì¼ ë¡œë“œ
naver_review = pd.read_csv("./db_data/ë±…ë±…ë§‰êµ­ìˆ˜_reviews.csv")
kakao_review = pd.read_csv("./db_data/bangbang_makguksu_reviews_kakao.csv")
google_review = pd.read_csv("./db_data/bangbang_makguksu_reviews_google.csv")

# MySQL ì—°ê²° ì„¤ì •
db = "mysql"
dbtype = "pymysql"
user = "root"          # MySQL ì‚¬ìš©ì ì´ë¦„
password = "1234"      # MySQL ë¹„ë°€ë²ˆí˜¸
host = "127.0.0.1:3306" # MySQL í˜¸ìŠ¤íŠ¸ì™€ í¬íŠ¸
database = "review_db" # ì‚¬ìš©í•  ë°ì´í„°ë² ì´ìŠ¤ ì´ë¦„

# SQLAlchemy ì—”ì§„ ìƒì„±
engine = create_engine(f"{db}+{dbtype}://{user}:{password}@{host}/{database}")

# ë°ì´í„°ë² ì´ìŠ¤ ìƒì„± (í•„ìš” ì‹œ)
with engine.connect() as conn:
    conn.execute("CREATE DATABASE IF NOT EXISTS review_db")
    conn.execute("USE review_db")

# ë°ì´í„°ë¥¼ MySQL í…Œì´ë¸”ì— ì €ì¥
def save_to_mysql(df, table_name):
    df.to_sql(
        table_name,
        con=engine,
        if_exists="replace",  # í…Œì´ë¸”ì´ ì´ë¯¸ ì¡´ì¬í•˜ë©´ ë®ì–´ì“°ê¸°
        index=False           # DataFrameì˜ ì¸ë±ìŠ¤ ì œì™¸
    )

# í…Œì´ë¸”ì— ë°ì´í„° ì €ì¥
save_to_mysql(naver_review, "naver_reviews")
save_to_mysql(kakao_review, "kakao_reviews")
save_to_mysql(google_review, "google_reviews")

print("ë°ì´í„° ì €ì¥ ì™„ë£Œ!")
```
