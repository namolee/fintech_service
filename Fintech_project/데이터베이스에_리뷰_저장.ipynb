```python
import pandas as pd
from sqlalchemy import create_engine

# CSV 파일 로드
naver_review = pd.read_csv("./db_data/뱅뱅막국수_reviews.csv")
kakao_review = pd.read_csv("./db_data/bangbang_makguksu_reviews_kakao.csv")
google_review = pd.read_csv("./db_data/bangbang_makguksu_reviews_google.csv")

# MySQL 연결 설정
db = "mysql"
dbtype = "pymysql"
user = "root"          # MySQL 사용자 이름
password = "1234"      # MySQL 비밀번호
host = "127.0.0.1:3306" # MySQL 호스트와 포트
database = "review_db" # 사용할 데이터베이스 이름

# SQLAlchemy 엔진 생성
engine = create_engine(f"{db}+{dbtype}://{user}:{password}@{host}/{database}")

# 데이터베이스 생성 (필요 시)
with engine.connect() as conn:
    conn.execute("CREATE DATABASE IF NOT EXISTS review_db")
    conn.execute("USE review_db")

# 데이터를 MySQL 테이블에 저장
def save_to_mysql(df, table_name):
    df.to_sql(
        table_name,
        con=engine,
        if_exists="replace",  # 테이블이 이미 존재하면 덮어쓰기
        index=False           # DataFrame의 인덱스 제외
    )

# 테이블에 데이터 저장
save_to_mysql(naver_review, "naver_reviews")
save_to_mysql(kakao_review, "kakao_reviews")
save_to_mysql(google_review, "google_reviews")

print("데이터 저장 완료!")
```
