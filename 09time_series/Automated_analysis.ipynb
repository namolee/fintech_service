```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import ta
import koreanize_matplotlib
import autogluon.timeseries as agts
from autogluon.timeseries import TimeSeriesDataFrame
from autogluon.timeseries import TimeSeriesPredictor
```

```python
aapl = yf.download('AAPL', start='2014-01-01', end='2024-10-16')
```

```python
# 'Close' 열만 사용하여 새로운 DataFrame 생성
aapl_close = aapl[['Close']].reset_index()

# AutoGluon의 TimeSeriesDataFrame 형식에 맞게 'item_id'와 'timestamp' 열을 추가
# item_id는 주식 이름을 사용하고, timestamp는 날짜로 설정
aapl_close['item_id'] = 'AAPL'
aapl_close.rename(columns={'Date': 'timestamp', 'Close': 'value'}, inplace=True)

# 모델 학습: freq 파라미터 추가
predictor = TimeSeriesPredictor(prediction_length=30, target='value', freq='D')

# 학습 진행
predictor.fit(train_data=aapl_close)

# 30일 주가 예측
future_forecast = predictor.predict(aapl_close)

# 결과 출력
print(future_forecast)
```

```python
print(future_forecast.head())
```

```python
# 실제 주가 데이터 (기존 Close 데이터를 사용)
actual_close = aapl[['Close']].reset_index()

# 예측 결과 생성 (이미 예측된 결과가 있다면 이 과정은 생략 가능)
future_forecast = predictor.predict(aapl_close)

# 예측된 주가 데이터 그래프 (주황색)
# 예측 결과의 타임스탬프는 실제 데이터의 마지막 날짜에서부터 이어집니다
future_dates = future_forecast.index.get_level_values('timestamp')

# 실제 데이터 그래프 (파란색)
plt.figure(figsize=(10, 6))
plt.plot(actual_close['Date'], actual_close['Close'], label='Actual Close Price', color='blue')

# 예측 데이터의 'mean' 열을 사용해 예측 데이터 그래프 (주황색)
plt.plot(future_dates, future_forecast['mean'], label='Predicted Close Price', color='orange')

# 그래프 제목 및 레이블 설정
plt.title('Apple Stock Price: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()

# 그래프 보여주기
plt.show()
```

# 자동화

```python
import yfinance as yf
import gradio as gr
import matplotlib.pyplot as plt
from autogluon.timeseries import TimeSeriesPredictor

# 주식 데이터 다운로드 함수
def get_stock_data(ticker):
    stock_data = yf.download(ticker, start='2014-01-01', end='2024-10-16')
    stock_close = stock_data[['Close']].reset_index()
    stock_close['item_id'] = ticker
    stock_close.rename(columns={'Date': 'timestamp', 'Close': 'value'}, inplace=True)
    return stock_close, stock_data

# 주가 예측 및 시각화 함수
def predict_and_plot(stock_name):
    # 주가 데이터 불러오기
    aapl_close, actual_close = get_stock_data(stock_name)
    
    # AutoGluon 모델 학습
    predictor = TimeSeriesPredictor(prediction_length=30, target='value', freq='D')
    predictor.fit(train_data=aapl_close)

    # 30일 주가 예측
    future_forecast = predictor.predict(aapl_close)

    # 예측된 주가 데이터 그래프
    future_dates = future_forecast.index.get_level_values('timestamp')

    # 그래프 그리기
    plt.figure(figsize=(10, 6))
    plt.plot(actual_close.index, actual_close['Close'], label='Actual Close Price', color='blue')  # 'timestamp' 대신 index로 처리
    plt.plot(future_dates, future_forecast['mean'], label='Predicted Close Price', color='orange')
    plt.title(f'{stock_name} Stock Price: Actual vs Predicted')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    
    # 그래프를 이미지로 변환
    plt.savefig('stock_prediction.png')
    plt.close()

    return actual_close.head(), actual_close.tail(), 'stock_prediction.png'

# Gradio 인터페이스 설정
with gr.Blocks() as app:
    gr.Markdown("## 주식 예측 및 시각화")
    
    # 드롭다운으로 선택할 주식 리스트
    stock_dropdown = gr.Dropdown(choices=['AAPL', 'MSFT', 'TSLA', 'NVDA'], label="주식을 선택하세요", value=None)
    
    # .head()와 .tail() 데이터 출력 및 그래프 표시
    head_output = gr.Dataframe(label="head()")
    tail_output = gr.Dataframe(label="tail()")
    graph_output = gr.Image(label="예측 그래프")

    # 분석중 표시 추가
    status = gr.Markdown(value="")

    # 버튼 클릭 시 예측 및 결과 출력
    predict_button = gr.Button("예측하기")

    def run_prediction(stock_name):
        # 분석중 표시
        status.value = "분석중..."
        
        # 예측 및 그래프 생성
        head, tail, graph = predict_and_plot(stock_name)
        
        # 분석 완료 후 결과 업데이트
        status.value = "분석 완료!"
        return head, tail, graph, status.value

    predict_button.click(run_prediction, inputs=stock_dropdown, outputs=[head_output, tail_output, graph_output, status])

# 앱 실행
app.launch()
```
