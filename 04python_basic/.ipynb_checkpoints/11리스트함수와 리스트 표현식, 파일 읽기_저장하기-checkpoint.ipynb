## 리스트 표현식(list comprehension) 사용하기
* 리스트 안에 for반복문과 if else 조건문을 한 번에 적는 방법
* 여러 줄의 코드를 1줄로 줄일 수 있고
* for, if else 문을 쓸 때 보다 처리속도가 빠르다
* [실행할 명령 for 변수 in list, tuple]

```python
list(range(1,11))
```

```python
a=[]
for i in range(1,11):
    a.append(i)

print(a)
```

```python
import time
```

```python
start=time.time()

a=[]
for i in range(1,20000000):
    a.append(i+5)
print(a[:10])
print(time.time()-start)
```

```python
start=time.time()

b=[i+5 for i in range(1,20000000)]
print(b[:10])
print(time.time()-start)
```

```python
import numpy as np
```

```python
start=time.time()
c=np.arange(1,200000000)
c=c+5
print(c[:10])
print(time.time()-start)
```

## 파일 읽기, 쓰기 open(), with open()

## 1) 파일에 문자열을 쓸 때는 open 함수로 파일을 열어서 파일 객체(file object)를 얻은 뒤에 write 메서드를 사용해 저장

(1) 파일 저장하기 <br>
변수명 = open(파일이름, 파일모드(w:쓰기, r:읽기) <br>
변수명.write('저장할 문자열') <br>
변수명.close()   #close()를 안하면 계속 열려 있음.

```python
file = open('test.txt','w') # test.txt로 파일 만들기
file.write('파일 만들기 연습') # test.txt에 .write()으로 내용이 채워진다.
file.close()
```

(2) 파일 읽기 <br>
변수명 = open(파일이름, 파일모드('r'읽기)) <br>
변수명2.read() <br>
변수명.close() 

```python
file2=open('test.txt', 'r')
my_file=file2.read()
file2.close()
print(my_file)
```

## with 문으로 파일 자동으로 닫기
* 파일을 만들려면 파일 만들기 => 내용 적기 => 파일닫기(저장) <br>
* 이때 close()를 안하면 오류 발생 <br>
* 이런 실수를 줄이기 위해서는 파일을 자동으로 닫아주는 with문을 사용하면 편리

```python
with open('test_with.txt','w') as file3:
    file3.write('with open을 사용하면 파일을 자동으로 닫아주어 편리하다.')
```

```python
with open('test_with.txt','r') as file4:
    text=file4.read()
print(text)
```

## encoding에 따른 읽기 오류
* 운영체제나(windows, linux, macOS), 한글/영어 윈도우에 따라서 인코딩 되는 포멧이 달라진다 <br>
* 인코딩에 따른 오류가 발생할 수 있음<br>
* web 표준은 utf-8 이므로 인코딩은 utf-8로 하는 것이 좋다.
* windows의 경우 인코딩은 보통 cp949로 되므로 주의

```python
with open('test_with2.txt','w', encoding='utf-8') as file5:
    file5.write("이것은 with open으로 만든 텍스트 파일이고 utf-8로 인코딩 함")
```

```python
with open("test_with2.txt",'r') as file6:
    file6_text=file6.read()
print(file6_text)
```

```python
with open("test_with2.txt",'r',encoding='utf-8') as file6:
    file6_text=file6.read()
print(file6_text)
```

```python
with open('test_with2.txt','a', encoding='utf-8') as file5:
    file5.write("이것은 with open으로 만든 텍스트 파일이고 utf-8로 인코딩 함2")
```

```python
with open("test_with2.txt",'r',encoding='utf-8') as file6:
    file6_text=file6.read()
print(file6_text)
```

```python
with open("./data/서울특별시_관광지입장정보_2011_2016.json",'r',encoding='utf-8') as f:
    tour_data=f.read()
print(tour_data)  
```

```python
data=json.loads(tour_data)
print(type(data))
data
```

```python
len(data)
```

```python
data[0]
```

```python
for item in data[:1]:
    #print((item.values()))
    for value in item.values():
        print(value)
```

```python
for items in data[:2]:
    for i in items:
        print(i,":",items[i])
```

```python
for item in data[:1]:
    for key, value in item.items():
        print(key,':',value)
```

```python
for items in data[:2]:
    for i in items:
        print(i,":",items.get(i,'none'))
```

```python
keylist=[]
for item in data[:1]:
    for key in item:
        keylist.append(key)
print(keylist)
```

```python
ForNumlist=[]
i=0
while i<67:
    ForNumlist.append(data[i]['ForNum'])
    i+=1
print(ForNumlist)
```

```python
datadic = {}
for datum in data:
    for k, v in datum.items():
        datadic.setdefault(k, []).append(v)
print(datadic)
```

```python
data[0].keys()
```

```python
ForNum = []
NatNum = []
addrCd = []
gungu = []
resNm = []
rnum = []
sido = []
yyyymm = []
for item in data[:]:
    #print(item['ForNum'])
    ForNum.append(item['ForNum'])
    NatNum.append(item['NatNum'])
    addrCd.append(item['addrCd'])
    gungu.append(item['gungu'])
    resNm.append(item['resNm'])
    rnum.append(item['rnum'])
    sido.append(item['sido'])
    yyyymm.append(item['yyyymm'])
```

```python
result={'ForNum':ForNum, 'NatNum':NatNum, 'addrCd':addrCd, 'gungu':gungu, 'resNm':resNm, 'rnum':rnum, 'sido':sido, 'yyyymm': yyyymm}
print(result)
```

```python
result2={}
for key in list(data[0].keys()):
    item_list=[]
    for item in data:
        item_list.append(item[key])
    result2[key]=item_list
print(result2)
```

```python
result2={}
for key in list(data[0].keys()):
    item_list=[]
    for item in data:
        item_list.append(item[key])
    result2[key]=item_list
print(result2)
```

```python
result3={}
for key in list(data[0].keys()):
    result3[key]=[item[key] for item in data]
print(result3)
```

# 리스트 표현식에 조건문 포함하기


```python
[i for i in range(10) if i%2==0]
```

```python
L=list(range(20))
```

```python
for i in L:
    if i%2==1:
        print("odd")
    else:
        print("even")
```

```python
for i in L:
    print("odd") if i%2==1 else print("even")
```
