```python
education_offices = [
    '경기도교육청', '경상남도교육청', '경상북도교육청', '전북특별자치도교육청',
    '광주광역시교육청', '대구광역시교육청', '전라남도교육청', '대전광역시교육청', '충청남도교육청',
    '서울특별시교육청', '인천광역시교육청', '세종특별자치시교육청', '울산광역시교육청', '부산광역시교육청',
    '충청북도교육청', '제주특별자치도교육청'
]

for office in education_offices:
    # 특정 교육청에 해당하는 데이터만 필터링
    filtered_data = data[data['시도교육청명'] == office]
    
    # 파일 이름 생성 (예: gangwon.csv)
    filename = office + '.csv'
    
    # CSV 파일로 저장
    filtered_data.to_csv(filename, index=False)
    print(f'{filename} 저장 완료.')
```
