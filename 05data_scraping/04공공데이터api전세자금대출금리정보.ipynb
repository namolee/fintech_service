```python
cloth_bins=[]
page=1
while True:
    url="https://api.odcloud.kr/api/15127131/v1/uddi:a9873b46-9551-407a-aff5-a3a77befb3d4"
    payload={'page':1,'perPage':10,'returnType':'JSON'}

    r = requests.get(url, params=payload)

    print(r.status_code)
    data=r.json()

    cloth_bins.append(data['data'])
    total_pages=data['totalCount']//data['perPage']+1
    print(f'{page}/{total_pages}',end='\r')
    if(page <total_pages):
        page+=1
    else:
        break
        
result = pd.DataFrame()
for bins in cloth_bins:
    temp=pd.json_normalize(bins)
    result=pd.concat([result,temp])
result
```

```python
chicken_list=[]
page=1
while True:
    url="http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    servicekey=os.getenv('servicekey')
    payload={'servicekey':servicekey,'divId':'indsSclsCd','key':'I21006','numOfRows':100,'pageNo':page,'type':'json'}

    r = requests.get(url, params=payload)
    data=r.json()

    chicken_list.append(data['body']['items'])
    total_pages=data['body']['totalCount']//data['body']['numOfRows']+1
#     print(f'{page}/{total_pages}',end='\r')
    if(page <total_pages):
        page+=1
    else:
        break

result = pd.DataFrame()
for chicken in chicken_list:
    temp=pd.json_normalize(chicken)
    result=pd.concat([result,temp])
result
```

## 서민금융상품기본정보

```python
fin_product=[]
page=1
while True:
    url="http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
    servicekey=os.getenv('serviceKey')
    payload={'serviceKey':servicekey,'numOfRows':10,'pageNo':page,'resultType':'json'}

    r = requests.get(url, params=payload)
    data=r.json()

    fin_product.append(data['response']['body']['items']['item'])
    total_pages=data['response']['body']['totalCount']//data['response']['body']['numOfRows']+1
    print(f'{page}/{total_pages}',end='\r')
    if(page <total_pages):
        page+=1
    else:
        break

result = pd.DataFrame()
for product in fin_product:
    temp=pd.json_normalize(product)
    result=pd.concat([result,temp])
result
```
