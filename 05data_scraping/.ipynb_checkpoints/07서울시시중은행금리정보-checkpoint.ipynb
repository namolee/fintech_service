```python
import requests
import os
from dotenv import load_dotenv
import pandas as pd
import time
from bs4 import BeautifulSoup
load_dotenv()
```

```python

url=f"http://openapi.seoul.go.kr:8088/{os.getenv('seoulkey')}/xml/dreamMoney/1/1000/"

r=requests.get(url)

data=r.text
soup = BeautifulSoup(data, 'xml')
keys={}
for item in soup.find_all('row'):
    for tag in item:
        if tag.name==None:
            continue
        else:
            keys.setdefault(tag.name,"")
result={}        
for key in keys:
    for item in soup.find_all(key):
        result.setdefault(item.name,[]).append(item.string)
result
result_df=pd.DataFrame(result)
result_df
```
