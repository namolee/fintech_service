```python
data = data.drop(['TotRmsAbvGrd', 'GarageArea', 'Foundation_PConc', 'BsmtQual_TA', 'GarageYrBlt'], axis=1)
data.head()
```

```python
X = data.drop(['SalePrice'],axis=1)
y = data['SalePrice']
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
import joblib
```

# joblib으로 스케일러 저장하기

```python
rs = RobustScaler()
X_scaled = rs.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
joblib.dump(rs, './model/iowa_robustscaler.joblib')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y , test_size=0.3 , random_state=10)
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model = Sequential()
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
early_stop = EarlyStopping(monitor='val_loss', patience=20)
modelpath = "./model/iowa_house{epoch:03d}__{val_loss:.4f}.keras"
checkpoint = ModelCheckpoint(filepath=modelpath, monitor='val_loss', save_best_only=True)
histroy = model.fit(X_train,y_train,epochs=2000, batch_size=200, validation_data=(X_test,y_test),callbacks=[early_stop, checkpoint])
```

```python
pred = model.predict(X_test)
pred = pd.DataFrame(pred, columns=['pred'])
pred
```

```python
y_test = pd.DataFrame(y_test)
y_test = y_test.reset_index(drop=True)
y_test
```

```python
final_result = y_test.join(pred)
final_result
```

```python
final_result
```

```python
plt.figure(figsize=(10, 5))
plt.plot(final_result['SalePrice'], label='SalePrice')
plt.plot(final_result['pred'], label='pred')
plt.legend()
plt.show()
```

# 트리계열의 알고리즘을 사용해 선분석 후 중요 변수만 추출해서 재분석

```python
data2.info()
```

```python
X2 = data2.drop('SalePrice', axis=1)
y2 = data2['SalePrice']
```

```python
X2_dummy = pd.get_dummies(X2).copy()
```

```python
X2_dummy.info(max_cols=300)
```

```python
rs2 = RobustScaler()
X2_scaled = rs2.fit_transform(X2_dummy)
X2_scaled = pd.DataFrame(X2_scaled, columns=X2_dummy.columns)
X2_scaled
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2_scaled, y2, test_size=0.3, random_state=10)
```

```python
from xgboost import XGBRegressor
```

```python
xgbr = XGBRegressor(max_depth=5, n_estimator=1000, n_jobs=-1)
xgbr.fit(X2_train, y2_train)
pred2 = xgbr.predict(X2_test)
print("mean_squared_error", mean_squared_error)
print("root_mean_squared_error", np.sqrt(mean_squared_error(y_test,pred2)))
print("r-squared", r2_score(y2_test, pred2))
```

```python
joblib.dump(xgbr, './model/iowa_xgbr.joblib')
```

```python
xgbr_result = pd.DataFrame(xgbr.feature_importances_, index=xgbr.feature_names_in_).sort_values(by=0, ascending=False)
```

```python
xgbr_result.head(100)
```

```python
xgbr_result['cumsum'] = xgbr_result.cumsum()
```

```python
xgbr_selected_col = xgbr_result[xgbr_result['cumsum'] <= 0.80 ].index
```

```python
xgbr_selected_col=list(xgbr_selected_col)
```

```python
xgbr_X = X2_scaled[xgbr_selected_col]
```

# XGBoost로 선별한 변수로 딥러닝 하기

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(xgbr_X,y2 , test_size=0.3, random_state=10)
```

```python
inputs = Input(shape=X3_train.shape[1])
x = Dense(32, activation='relu')(inputs)
x = Dense(64, activation='relu')(x)
x = Dense(128, activation='relu')(x)
x = Dense(256, activation='relu')(x)
x = Dense(32, activation='relu')(x)
x = Dense(1)(x)
```

```python
model2 = Model(inputs, x)
model2.summary()
```

```python
model2.compile(loss='mean_squared_error',optimizer='adam' )
early_stop2 = EarlyStopping(patience=100)
modelpath="./model/xgbselected_{epoch:04d}__{val_loss:.4f}.keras"
checkpoint2 = ModelCheckpoint(filepath=modelpath, monitor='val_loss', save_best_only=True)
history2 = model2.fit(X3_train, y3_train, epochs=2000, batch_size=800, callbacks=[early_stop2,checkpoint2],
                     validation_data=(X3_test, y3_test))
```
