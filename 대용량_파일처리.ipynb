```python
import pandas as pd
import glob

# 파일 경로 설정
path = './mini_project/아파트매매실거래가1999-202408/*.csv'  # 해당 폴더에 있는 모든 csv 파일을 선택
all_files = glob.glob(path)

# 모든 파일을 읽어 하나의 리스트에 저장
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, encoding='cp949', 
                     skiprows=15)  # 처음 15줄을 건너뛰고 16번째 줄부터 읽기
    li.append(df)

# 리스트에 저장된 모든 데이터프레임을 하나로 합치기
apt_data = pd.concat(li, axis=0, ignore_index=True)

# 결과 확인
print(apt_data)
```

### 금리 데이터 가져오기

```python
interest_data = pd.read_csv("./mini_project/기준금리1999-2023.csv")
interest_data
```

```python
# 변경일자와 Unnamed: 1 컬럼을 문자열로 변환하여 합치기
interest_data['변경일자'] = interest_data['변경일자'].astype(str)
interest_data['Unnamed: 1'] = interest_data['Unnamed: 1'].str.replace(' ', '')  # 공백 제거

# 변경일자와 월일 데이터를 합쳐서 새로운 '변경날짜' 컬럼 생성
interest_data['변경날짜'] = interest_data['변경일자'] + '-' + interest_data['Unnamed: 1']

# 문자열을 날짜 형식으로 변환
interest_data['변경날짜'] = pd.to_datetime(interest_data['변경날짜'], format='%Y-%m월%d일')

# 결과 확인
print(interest_data[['변경날짜']].head())
```

```python
interest_data.head()
```

```python
interest_data = interest_data[['변경날짜', '기준금리']]
```

```python
interest_data.head()
```

```python
# 계약날짜와 변경날짜의 타입이 datetime으로 되어 있는지 확인하고, 필요시 변환
apt_data['계약날짜'] = pd.to_datetime(apt_data['계약날짜'])
interest_data['변경날짜'] = pd.to_datetime(interest_data['변경날짜'])

# 기준금리를 계약날짜에 맞춰서 apt_data에 추가 (가장 가까운 과거 기준금리 매칭)
apt_data = pd.merge_asof(apt_data.sort_values('계약날짜'),
                         interest_data.sort_values('변경날짜'),
                         left_on='계약날짜',
                         right_on='변경날짜',
                         direction='backward')

# 결과 확인
apt_data.head()
```

```python
# apt_data의 시군구에서 앞 3글자를 추출
apt_data['지역코드'] = apt_data['시군구'].str[:3]

# 지역에 따른 시도교육청명을 매칭하는 함수 정의
def get_education_office(region):
    if '서울' in region:
        return '서울특별시교육청'
    elif '부산' in region:
        return '부산광역시교육청'
    elif '대구' in region:
        return '대구광역시교육청'
    elif '인천' in region:
        return '인천광역시교육청'
    elif '광주' in region:
        return '광주광역시교육청'
    elif '대전' in region:
        return '대전광역시교육청'
    elif '울산' in region:
        return '울산광역시교육청'
    elif '세종' in region:
        return '세종특별자치시교육청'
    elif '경기' in region:
        return '경기도교육청'
    elif '강원' in region:
        return '강원특별자치도교육청'
    elif '충청북' in region:
        return '충청북도교육청'
    elif '충청남' in region:
        return '충청남도교육청'
    elif '전북' in region or '전라북' in region:
        return '전북특별자치도교육청'
    elif '전라남' in region:
        return '전라남도교육청'
    elif '경상북' in region:
        return '경상북도교육청'
    elif '경상남' in region:
        return '경상남도교육청'
    elif '제주' in region:
        return '제주특별자치도교육청'
    else:
        return None

# 시도교육청명 컬럼 추가
apt_data['시도교육청명'] = apt_data['지역코드'].apply(get_education_office)

# 지역코드 컬럼 제거 (필요시)
apt_data.drop(columns=['지역코드'], inplace=True)

# 결과 확인
print(apt_data[['시군구', '시도교육청명']].head())
```

## 학원 데이터 불러오기

```python
# 1. academy_data에서 시도교육청명과 행정구역명별 학원 수 계산
academy_counts = academy_data.groupby(['시도교육청명', '행정구역명']).size().reset_index(name='학원수')

# 2. apt_data에 학원수 컬럼을 기본값 None으로 설정
apt_data['학원수'] = None

# 3. 행정구역명에서 '시', '군', '구' 제거 (정규 표현식을 이용)
academy_counts['행정구역명_clean'] = academy_counts['행정구역명'].str.replace(r'(시|군|구)$', '', regex=True)

# 4. 벡터화된 방식으로 시군구와 행정구역명을 비교
for i, region_row in academy_counts.iterrows():
    # 시군구에 해당 행정구역명이 포함되어 있고, 시도교육청명도 일치하는지 확인
    mask = (apt_data['시도교육청명'] == region_row['시도교육청명']) & apt_data['시군구'].str.contains(region_row['행정구역명_clean'])
    # 해당하는 행에 학원수 추가
    apt_data.loc[mask, '학원수'] = region_row['학원수']

# 결과 확인
print(apt_data[['시군구', '학원수', '시도교육청명']].head())
```

```python
import pandas as pd

# 1. 날짜 컬럼을 datetime 형식으로 변환
rate_data['date'] = pd.to_datetime(rate_data['date'])

# 2. 전체 날짜 범위 생성 (1995-01-03 ~ 2024-09-25)
date_range = pd.date_range(start='1995-01-03', end='2024-09-25')

# 3. 기존 데이터를 이 날짜 범위에 맞춰 재구성 (누락된 날짜를 추가)
rate_data_full = pd.DataFrame(date_range, columns=['date']).merge(rate_data, on='date', how='left')

# 4. 환율 값이 0이거나 비어 있는 경우 이전 값으로 채우기
rate_data_full['현찰_살때_환율'].replace(0, pd.NA, inplace=True)  # 0인 값들을 NA로 처리
rate_data_full['현찰_살때_환율'].fillna(method='ffill', inplace=True)  # 이전 값으로 채우기

# 결과 확인
rate_data_full.head()
```

```python
rate_data_full
```

```python
# 1. apt_data의 계약날짜 컬럼과 rate_data_full의 date 컬럼이 datetime 형식인지 확인
apt_data['계약날짜'] = pd.to_datetime(apt_data['계약날짜'])
rate_data_full['date'] = pd.to_datetime(rate_data_full['date'])

# 2. apt_data와 rate_data_full을 계약날짜(date)를 기준으로 병합
apt_data = pd.merge(apt_data, rate_data_full[['date', '현찰_살때_환율']], 
                    how='left', 
                    left_on='계약날짜', 
                    right_on='date')

# 3. 달러환율 컬럼 이름 변경 (현찰_살때_환율을 달러환율로 변경)
apt_data.rename(columns={'현찰_살때_환율': '달러환율'}, inplace=True)

# 4. 병합 후 필요 없는 date 컬럼 삭제
apt_data.drop(columns=['date'], inplace=True)

# 결과 확인
print(apt_data[['계약날짜', '달러환율']].head())
```
